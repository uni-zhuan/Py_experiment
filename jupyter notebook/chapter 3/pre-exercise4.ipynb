```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
colName=["学号","姓名","专业"]
data1=[["19000000","张三","智科"],["19000001","李四","应数"],["19000002","王五","人工智能"]]
stu1=DataFrame(data1,columns=colName)
data2=[["19000003","张六","智科"],["19000004","李七","应数"],["19000005","王八","人工智能"]]
stu2=DataFrame(data2,columns=colName)
```

```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
data=pd.read_excel("F:\Python\data\DataScience.xls")
data
```

```python
print(data.index)
print(data.columns)
```

```python
print(data.isnull().any())
```

```python
pre=data[data.isnull().T.any()]
pre.to_csv("F:\Python\data\pre.csv")
```

```python
data.loc[:,["课程名称","实验项目名称","实验类型","二级实验室名称"]]
```

```python
classNum=data.groupby(["课程名称"]).aggregate({"实验课时数":np.sum})
print(classNum)
```

```python
weekClassNum=data.groupby(["星期"]).aggregate({"实验课时数":np.sum})
print(weekClassNum)
```

```python
pd.crosstab(data["课程名称"],data["实验类型"])
```

```python
dataClass=data[["班级","课程名称","星期","周次","节次","实验地点门牌号"]]
dataClass
```

```python
classTab=dataClass.groupby(["课程名称"])
print(classTab)
```

```python
labNum=data.groupby(["二级实验室名称"]).aggregate({"实验课时数":np.sum})
labNum
```

```python
labType=data.groupby(["二级实验室名称","实验类型"])
labType
```

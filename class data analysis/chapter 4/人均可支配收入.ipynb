```python
import matplotlib.pyplot as plt
from pandas import Series,DataFrame
RMB=[1.47, 1.62, 1.78, 1.94, 2.38, 2.60]
plt.plot(RMB,linestyle="dashed",color="red",marker="s",label="Income")
plt.title("Income chart")
plt.grid()
plt.xlim(0,5)
plt.ylim(0,3.0)
plt.xticks((0,6),("2012","2013","2014","2015","2016","2017"))
plt.legend(loc="upper right")
plt.annotate("Largest!",xy=(5,2.6),xytext=(3,2.6),arrowprops=dict(arrowstyle="->"))
plt.xlabel("Year",fontsize=12)
plt.ylabel("Income(RMB Ten Thousand)",fontsize=12)
plt.savefig("F:\Python\data\Income.jpg",dpi=200,bbox_inches="tight")
plt.show()
```

```python
RMB=[1.47, 1.62, 1.78, 1.94, 2.38, 2.60]
data=DataFrame({"Income":RMB},index=[2012,2013,2014,2015,2016,2017])
data.plot(title="Income chart",ylim=(0,3),xlim=(2012,2017),color="red",linestyle="dashed",marker="s")
plt.xlabel("Year",fontsize=12)
plt.ylabel("Income(RMB Ten Thousand)",fontsize=12)
plt.savefig("F:\Python\data\Income.jpg",dpi=200,bbox_inches="tight")
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
plt.subplots_adjust(wspace=0.3,hspace=0.3)
ax1=fig.add_subplot(2,2,1)
data.plot(ax=ax1,title="line chart")
plt.ylabel("Income")
ax1.set_xlabel("Year")
ax2=fig.add_subplot(2,2,2)
data.plot(kind="box",fontsize="small",xticks=[],ax=ax2,title="box chart")
plt.xlabel("2012-2017")
ax3=fig.add_subplot(2,1,2)
data.plot(kind="bar",fontsize="medium",yticks=[0.0,1.0,2.0,3.0],ax=ax3,title="bar chart")
ax=["ax1","ax2","ax3"]
plt.show()
```

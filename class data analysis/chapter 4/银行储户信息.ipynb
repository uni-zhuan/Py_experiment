```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from pandas import Series,DataFrame
data=pd.read_csv("F:\Python\data\Bankpep.csv")
data["age"].plot(kind="hist",bins=10,title="Customer Age",normed=True)
data["age"].plot(kind="kde",ylim=[0.000,0.025],title="Customer Age")
plt.show()
```

```python
data.plot(kind="scatter",x="age",y="income",title="Customer Income",grid=True)
plt.show()
```

```python
condata=data[["age","income","children"]]
pd.scatter_matrix(condata,diagonal="hist")
plt.show()
```

```python
data1=data.groupby(["region"])
mean=data1.aggregate({"income":np.mean})
std=data1.aggregate({"income":np.std})
mean.plot(kind="bar",yerr=std,title="Customer Income",rot=45)
plt.show()
```

```python
fig=plt.figure(figsize=(4,4))
plt.subplots_adjust(wspace=0.5,hspace=0.5)
data0=data["sex"].groupby(data["sex"])
sex_count=data0.count()
sex_count
```

```python
data1=data[data["car"]=="YES"]
carsex=data1["sex"].groupby([data1["sex"]])
car_count=carsex.count()
car_count
```

```python
data2=data["children"].groupby(data["children"])
child_count=data2.count()
child_count
```

```python
ax1=fig.add_subplot(2,2,1)
sex_count.plot(kind="pie",autopct="%1.1f%%",ax=ax1,title="customer sex")

ax2=fig.add_subplot(2,2,2)
car_count.plot(kind="pie",autopct="%1.1f%%",ax=ax2,title="customer sex car")

ax3=fig.add_subplot(2,2,3)
child_count.plot(kind="pie",autopct="%1.1f%%",ax=ax3,title="customer children")
plt.show()
```

```python
import numpy as np
import pandas as pd 
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
x=np.linspace(0,6.28,50)#生成线性空间，start，end，num-points(越多越光滑)
y=np.sin(x)#计算y数组
plt.plot(x,y,color="red")#绘图函数
plt.plot(x,np.exp(-x),c="b")#另一种绘图
plt.show()
```

```python
sdata=pd.read_csv("F:\Python\data\students.csv")
sdata.plot(kind="scatter",x="Height",y="Weight",title="Students Body Shape",marker="*",xlim=[150,200],ylim=[40,80],label="(Height,Weight)")
plt.show()
```

```python
data1=sdata[sdata["Gender"]=="male"]
data2=sdata[sdata["Gender"]=="female"]
plt.figure()
plt.scatter(data1["Height"],data1["Weight"],c="b",marker="^",label="Female")
plt.scatter(data2["Height"],data2["Weight"],c="r",marker="^",label="Male")
plt.xlim(150,200)
plt.ylim(40,80)
plt.title("Students Body Shape")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid()
plt.legend(loc="upper right")
plt.show()
```

```python
data=pd.read_csv("F:\Python\data\population.csv",index_col="Year")
data1=data[["Boys","Girls"]]
```

```python
mean=np.mean(data1,axis=0)
std=np.std(data1,axis=0)
mean,std
```

```python
fig=plt.figure(figsize=(6,2))
plt.subplots_adjust(wspace=0.6,hspace=0.6)
ax1=fig.add_subplot(1,2,1)
mean.plot(kind="bar",yerr=std,color="cadetblue",title="Average of Births",rot=45)#文字倾斜，误差线用xerr
ax2=fig.add_subplot(1,2,2)
mean.plot(kind="barh",xerr=std,color="cadetblue",title="Average of Births")
data.plot(kind="bar",title="Births of Boys & Girls")
data.plot(kind="bar",stacked=True,title="Births of Boys & Girls")
plt.show()
```

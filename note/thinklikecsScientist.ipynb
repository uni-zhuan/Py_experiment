```python
#14.1
import os
def walk(dirname):
    for name in os.listdir(dirname):
        path=os.path.join(dirname, name)
        if os.path.isfile(path):
            print(path)
        else:
            walk(path)#递归

# walk('E:\study\Experiments\python_calu')
os.walk('E:\study\Experiments\python_calu')
help(os.walk)
```
